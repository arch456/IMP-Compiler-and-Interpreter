## Part I: Introduction

### The IMP Language 
IMP is a simple **Imperative** programming language. 
IMP programs have three syntactic entities:

> a) **Statements** 

> b) **Arithmetic expressions** (known as “aexps”)

> c) **Boolean expressions** (also “bexps”) 

- An IMP program can consist of a single statement, a list of statements, a block of statements or just an empty string. 
 
- Arithmetic expressions evaluate to integers and Boolean expressions evaluate to true(1) or false(0). 
 
- The data types supported by IMP language are Integers and Booleans. 
 


### Types of IMP Statements:
 
> **Assignment Statement:**
It sets a variable to an integer obtained by evaluating an aexp. 
Example:    a := 3 * 4 
 
> **Compound Statement:**
It consists of two sub-statements, separated by a semicolon. 
Multiple statements may be linked together using this. 
Example:   a := 5;             b := 4;             c := 6 
 
> **If-then-else Statement:**
It is made up of a bexp condition, and two sub-statements, followed by an optional *else* statement. 
If the condition evaluates to true, the first statement gets executed. Otherwise, the second statement is executed. 
Example:   
```
if a < b then             
                print a; 
           else             
                print b; 
           end 
``` 
>**While Statement:**
It includes a bexp condition and a loop body sub-statement. 
If the condition evaluates to true, the loop body gets executed. This is repeated until the condition evaluates to false. Example:   
```
while x < 10 do 
    print x;
    x := x + 1;
end 
```
>**Input Statement:**
This statement is used to read input from the users

>**Print Statement:**
This is used to print output to the screen
 



### Types of Arithmetic Expressions (aexps)
 
> **Integer Expression:** 
Consists of an integer that just evaluates to itself. 
 ```
      48 
 ```
> **Variable Expression:** 
It gives the current value of a variable. 
If the variable has not been assigned, by default 0 is assigned to it.              

    ```
    a
    ```

    ```
    b := 4;
    ```
 
> **Binary Operator Expression:**
It is used to perform an arithmetic operation (+, -, *, /) on two aexps. 
 ```
 9 + a     
 ```
 ```
 b * 5 
 ```


### Types of Boolean expressions (bexps)
 
> **Relational Operator Expression:**
It is used to perform a relational operation (<, <=, >, >=, =, !=) on two sub-aexps. 
```
a < 10 
b >= 20
c == 30
```
>**NOT Expression:**
It contains a sub-bexp, prefixed by “not”. Evaluates to the opposite of whatever the sub-expression evaluates to. 
```
not d < 12 
```



### Java Bytecode

- Java bytecode is the instruction set of the **Java virtual machine (JVM)**. 

- The JVM is both a stack machine and a register machine. 

- Each frame for a method call has an **operand stack** and an array of **local variables**. 

- The operand stack is used for operands to computations and for receiving the return value of a called method, while local variables serve the same purpose as registers and are also used to pass method arguments. 

- The maximum size of the operand stack and local variable array, computed by the compiler, is part of the attributes of each method.

### Instruction set

Each bytecode is composed of one byte that represents the **opcode**, along with zero or more bytes for **operands**.
There are about 256 possible byte-long opcodes.

The Instructions that we have utilized in our implementation are:

>**Load and store** (for example: aload_0, istore)

>**Arithmetic** (for example: iadd, imul)

>**Logical** (for example: if_icmple, if_icmpgt)

>**Method invocation and return** (for example: invokevirtual, return)


### Translation to Bytecode
We look at an example to understand how the bytecode instructions get generated for an example program:

Program:
```
    int a = 1;
    int b = 2;
    int c = a + b;
```
Translating the above program to Java bytecode, we get:
```
0: iconst_1
1: istore_1
2: iconst_2
3: istore_2
4: iload_1
5: iload_2
6: iadd
7: istore_3
8: ireturn
```

Meaning of each instruction:
>*iconst_1*: load the int value 1 onto the stack

>*istore_1*: store int value into variable 1

>*iload_2* : load an int value from local variable 2

>*iadd*    : add two ints

>*ireturn*  : return value from a method

#### Description:
The following components constitute the IMP language implementation:

1. **Lexer**

Lexer converts the IMP program instructions that go as input to it, into a sequence of tokens (token stream). 
Tokens are strings with an assigned and thus identified meaning. The lexer in this implmentation has been built using the **lex** tool within Python's **PLY** package. It is specified in the source file **IMP_lex.py**.


2. **Parser (Frontend)**

Parser syntactically analyzes the token stream generated by the lexer and generates an Abstract Syntax Tree. ASTs show the syntactic relation of various components with each other and  may also contain semantic and other information. Syntactic analysis is a formal analysis of the input code into its component parts in order to facilitate the writing of compilers and interpreters. For this implementation, the **yacc** parsing tool within the **PLY** package. The source file specifying the grammar is **IMP_gram.py**. The frontend is a parser which generates the Abstract Syntax Tree and it is specified in **IMP_frontend_gram.py**


3. **Interpreter**

The Interpreter directly executes instructions written in the IMP programming language and produces the output.Interpreter does not require that the program be complied into an intermediate language program. 
As per the architecture specified above, the Interpreter takes the Abstract Syntax Tree (AST) produced by the Frontend as input. It then executes the instructions and generates the output of the program. The Interpreter is specified in the files **IMP_interp_walk.py** and **IMP_interp.py**.


4. **Compiler**

The Compiler translates code written in IMP programming language into Java bytecode, that can be executed by a Java Virtual Machine (JVM). The **IMP_compiler_codegen.py** specifies the compiler for IMP language. It takes as input the AST generated by the Frontend parser and produces the corresponding instruction set in Java bytecode.

**Other Components**

*IMP_state.py*: The state is used to initialize and define the structure of the implementation

*grammar_stuff.py*: This contains some support functions for the grammar

*IMP_output.py*: This consists of some functions used to produce the compiler output in the format we want.
